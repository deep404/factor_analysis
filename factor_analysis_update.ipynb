#### Adequacy Test

Before performing factor analysis, it needs to evaluate the “factorability” of the dataset. Factorability means "can it be found the factors in the dataset?". There are two methods to check the factorability or sampling adequacy:
- Bartlett's Test
- Kaiser-Mayer-Olkin Test

**Bartlett’s test** of sphericity checks whether or not the observed variables intercorrelate at all using the observed correlation matrix against the identity matrix. If the test found statistically insignificant, you should not employ a factor analysis.

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value=calculate_bartlett_sphericity(df)
chi_square_value, p_value
```

In this Bartlett ’s test, the p-value is 0. The test was statistically significant, indicating that the observed correlation matrix is not an identity matrix.

**Kaiser-Meyer-Olkin (KMO) Test** measures the suitability of data for factor analysis. It determines the adequacy for each observed variable and for the complete model. KMO estimates the proportion of variance among all the observed variable. Lower proportion id more suitable for factor analysis. KMO values range between 0 and 1. Value of KMO less than 0.6 is considered inadequate.

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df)
```

```python
kmo_model
```

The overall KMO for data is 0.9, which is excellent. This value indicates that it is possible to proceed with planned factor analysis.

#### Choosing the Number of Factors


For choosing the number of factors, it can used the Kaiser criterion and scree plot. Both are based on eigenvalues.

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer()
#fa.analyze(df, 25, rotation=None)
fa.fit(df)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev
```

Here, it can be seen only for 8-factors eigenvalues are greater than one. It means it is necessary to choose only 8 factors (or unobserved variables).

```python
# Create scree plot using matplotlib
plt.scatter(range(1,df.shape[1]+1),ev)
plt.plot(range(1,df.shape[1]+1),ev)
plt.axhline(y = 1, color = 'r', linestyle = '--')
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

The scree plot method draws a straight line for each factor and its eigenvalues. Number eigenvalues greater than one considered as the number of factors. Red line is delimites eigenvalue equal to 1.

Here, it can be seen only for 8-factors eigenvalues are greater than one. It means it is necessary to choose only 8 factors (or unobserved variables).

### Construction of ESG index

Calculate the normalised squared factor loadings

```python
# Square all values
esg = results_new.copy()
esg.iloc[:, 1:] = np.square(results_new.iloc[:, 1:])

# Environmental variables 0 -> 7
# Social variables 8 -> 39
# Governance variables 40 -> 48
 
# apply zeros to unnecessary cells
esg.iloc[: 8, 2] = 0 # for environmental variables
esg.iloc[8 : 40, 1] = 0 # for social variables
esg.iloc[40 : , 2] = 0 # for governance variables
esg.iloc[:, 3] = 0

# extract current elements for normalization
factor1_column = esg.iloc[: 8, 1].to_list() + esg.iloc[40 :, 1].to_list()
factor2_column = esg.iloc[8 : 40, 2].to_list()

# Normalize elements
factor1_column_updated = np.array([round(x / sum(factor1_column), 2) for x in factor1_column])
factor2_column_updated = np.array([round(x / sum(factor2_column), 2) for x in factor2_column])

# Update the results dataframe with normalised values
esg.iloc[: 8, 1] = factor1_column_updated[:8]
esg.iloc[40 : , 1] = factor1_column_updated[8:]
esg.iloc[8 : 40 , 2] = factor2_column_updated

# Display updated dataframe results
esg.to_excel('factors_phase_2.xlsx', index = False)
esg
```

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer()
fa.fit(esg.iloc[:, 1:3])
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
print('Eiganvalue for Factor 1 is', round(ev[0], 2), 'and for Factor 2 is ', round(ev[1], 2))
```

```python
print('Weight of factors in summary indicator for Factor 1 is ', round(ev[0], 2) / ( round(ev[0], 2) +  round(ev[1], 2)), ' and for Factor 2 is ', round(ev[1], 2) / ( round(ev[0], 2) +  round(ev[1], 2)))
```

### Plot Bond Spread by Country and Year

```python
# Read excel data for plot
df = pd.read_excel('bond_spread_data.xlsx')

# Select necessary columns
df = df[['Country', 'Income Group', 'SDG Index ', 'Year', 'Bond_Spread']]

# Transform column type to float
df = df.astype({'Bond_Spread':float})

# group cells by 'Income Group' and 'Year'
df.groupby(by = ['Income Group', 'Year'])['Bond_Spread'].mean()

# Select the unique 'Bond_Spread'
mean_bond_spread = df.groupby(by = ['Income Group', 'Year'])['Bond_Spread'].mean().to_list()

# plot the graph
fig, ax = plt.subplots(figsize = (16, 10))

ax.scatter(list(df.Year.unique()), mean_bond_spread[:5], color = 'red')
ax.plot(list(df.Year.unique()), mean_bond_spread[:5], color = 'red', label = 'High income countries')

ax.scatter(list(df.Year.unique()), mean_bond_spread[5:10], color = 'blue')
ax.plot(list(df.Year.unique()), mean_bond_spread[5:10], color = 'blue', label = 'Lower middle income countries')

ax.scatter(list(df.Year.unique()), mean_bond_spread[10:], color = 'green')
ax.plot(list(df.Year.unique()), mean_bond_spread[10:], color = 'green', label = 'Upper middle income countries')


plt.title('Mean Bond Spread of countries by Income Group and Year', fontsize = 24)
plt.xlabel('Year', fontsize = 18)
plt.ylabel('Mean Bond Spread', fontsize = 18)
plt.xticks(list(df.Year.unique()))
plt.legend()
plt.grid()
plt.show()

fig.savefig('bond_spread_countries_year.png')
```
